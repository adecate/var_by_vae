```python
# from scipy.stats import lognorm
import matplotlib.pyplot as plt
import scipy.stats as stats


def ticket_distr(df, ticket):
    fig, ax = plt.subplots(figsize=(10, 6))

    values = df[ticket]

    shape, loc, scale = stats.t.fit(values)
    x = np.linspace(values.min(), values.max(), len(values))
    pdf = stats.t.pdf(x, shape, loc=loc, scale=scale)
    label = 'mean=%.4f, std=%.4f, shape=%.4f' % (loc, scale, shape)

    ax.hist(values, bins=50, density =True,alpha=0.5)
    ax.plot(x, pdf, 'r-', lw=2, label=label)
    ax.legend(loc='best')
    ax.set_title("{} log-returns vs Student distribution".format(ticket))
    plt.savefig("{}.png".format(ticket))
    plt.show();
```

```python
for ticket in crypto.columns:
    ticket_distr(crypto[1:], ticket)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#     
data = np.random.pareto(a=2.62, size=1000)

#  Log-Log    
log_data = np.log(data)
sorted_data = np.sort(log_data)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
plt.plot(sorted_data, np.log(1 - cdf), label='Pareto Distribution')

#     
normal_data = np.random.normal(loc=0, scale=1, size=1000)

#  Log-Log     
log_normal_data = np.log(np.abs(normal_data))  #   ,       
sorted_normal_data = np.sort(log_normal_data)
normal_cdf = np.arange(1, len(sorted_normal_data) + 1) / len(sorted_normal_data)
plt.plot(sorted_normal_data, np.log(1 - normal_cdf), 'r--', label='Normal Distribution')

#    
plt.title("Log-Log Plot")
plt.xlabel("log(x)")
plt.ylabel("log(1 - CDF)")
plt.legend()
plt.show()
```

```python
weights = np.array([1/13]*13)
crypto_portf = pd.DataFrame()
crypto_portf['protfolio'] = crypto.dot(weights)
crypto_portf['diff'] = crypto_portf.protfolio - crypto_portf.protfolio.shift(1)
crypto_portf
```

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(crypto_portf['diff'][1:].values.squeeze(), ax=ax, zero=False)
plt.ylim(-0.75, 0.75)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(crypto_portf['diff'][1:].values.squeeze(), ax=ax, zero=False)
plt.ylim(-0.75, 0.75)
plt.show()
```
